## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
import pandas as pd
import datetime as dt
df_cadastro = pd.read_csv('input_files/cadastro.csv', sep =',') 
df_investimento_1 = pd.read_csv('input_files/investimento_parte1.csv', sep =',')
df_investimento_2 = pd.read_csv('input_files/investimento_parte2.csv', sep =',')
print(f"O dataframe 'cadastro' tem {df_cadastro.shape[0]} linhas")
#print(df_cadastro.info())
print(f"O dataframe 'investimento_parte1' tem {df_investimento_1.shape[0]} linhas")
#print(df_investimento_1.info())
print(f"O dataframe 'investimento_parte2' tem {df_investimento_2.shape[0]} linhas")
#print(df_investimento_2.info())
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
# Juntando os dois dataframes de investimentos, com as mesmas colunas, por meio das linhas.
df_investimento_total = pd.concat([df_investimento_1,df_investimento_2])

# Mostrando a quantidade de linhas do novo dataframe
df_investimento_total.shape[0]
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
# Verificando Linhas Duplicadas
df_investimento_total.duplicated()

# Eliminando as duplicadas (Ignorando os indices, mantendo a primeira ocorrência e sem criar subset)
df_investimento = df_investimento_total.drop_duplicates(subset = None, keep = 'first',ignore_index=True)
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
# Juntando os dataframes cadastro e investimento, colocando as colunas do investimento à direita e alinhando pelos 'id'
df_cadas_invest = pd.merge(df_cadastro, df_investimento, on ='id', how = 'right')
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
# Ordenando pelo Id
df_ordenado = df_cadas_invest.sort_values(['id'])

# Redefinindo os índices, dropando os índices atuais e substituindo pelos novos
df_ordenado.reset_index(drop=True,inplace = True)
```
